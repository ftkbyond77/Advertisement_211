```python
def reverse_normalize(tensor_img):
    """
    ย้อน normalize จาก [-1, 1] → [0, 1]
    รองรับ Tensor shape (C, H, W) หรือ (N, C, H, W)
    """
    return (tensor_img * 0.5) + 0.5
```
