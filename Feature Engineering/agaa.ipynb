```python
import json
import re
import pandas as pd
from collections import defaultdict

# Load the original data
with open('copter.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Function to clean text values
def clean_text(text_list):
    if not text_list:
        return []
    
    cleaned = []
    for item in text_list:
        # Remove special characters but keep Thai characters and English letters/numbers
        clean_item = re.sub(r'[^\w\s\.\,\'\"\-\u0e00-\u0e7f]', ' ', item)
        # Replace multiple spaces with a single space
        clean_item = re.sub(r'\s+', ' ', clean_item).strip()
        # Remove very short items (less than 2 characters)
        if len(clean_item) > 1:
            cleaned.append(clean_item)
    
    return cleaned

# Clean the data
cleaned_data = {}
for key, values in data.items():
    cleaned_data[key] = clean_text(values)

# Convert to a more analysis-friendly format
sentiment_df = pd.DataFrame([
    {"id": key, "text": " ".join(values)} 
    for key, values in cleaned_data.items() 
    if values  # Only include non-empty values
])

# Preview the cleaned data
print(f"Total records: {len(sentiment_df)}")
print("\nSample entries:")
print(sentiment_df.head(10))

# Export to CSV for sentiment analysis
sentiment_df.to_csv('cleaned_sentiment_data.csv', index=False)

# Additionally, create a format that preserves individual phrases per record
expanded_data = []
for key, values in cleaned_data.items():
    if values:
        for text in values:
            if text.strip():  # Only add non-empty strings
                expanded_data.append({"id": key, "text": text})

expanded_df = pd.DataFrame(expanded_data)
expanded_df.to_csv('expanded_sentiment_data.csv', index=False)

print(f"\nTotal individual phrases: {len(expanded_df)}")

# Group by language for potential separate model training
def detect_language(text):
    # Simple detection - if it contains Thai characters, classify as Thai
    if re.search(r'[\u0e00-\u0e7f]', text):
        return 'thai'
    return 'other'  # Mostly English or numbers

expanded_df['language'] = expanded_df['text'].apply(detect_language)
language_counts = expanded_df['language'].value_counts()
print("\nLanguage distribution:")
print(language_counts)

# Create dictionary of processed data for easy inspection
final_clean_data = defaultdict(list)
for _, row in expanded_df.iterrows():
    final_clean_data[row['id']].append(row['text'])

# Convert defaultdict to regular dict for JSON serialization
final_clean_dict = dict(final_clean_data)

# Pretty printed version of the cleaned data
pretty_json = json.dumps(final_clean_dict, ensure_ascii=False, indent=2)

# Print a small sample
print("\nSample of cleaned JSON structure:")
sample_keys = list(final_clean_dict.keys())[:3]
sample_data = {k: final_clean_dict[k] for k in sample_keys}
print(json.dumps(sample_data, ensure_ascii=False, indent=2))

# Return the fully cleaned data
print("\nFull cleaned data dictionary ready for sentiment analysis")
```

```python
import json
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
from collections import Counter

# Let's simulate loading the cleaned data we created
# In a real scenario, you would load from the CSV or JSON files generated
with open('copter.json', 'r', encoding='utf-8') as f:
    raw_data = json.load(f)

# Clean and prepare data for sentiment analysis
def prepare_sentiment_data(data):
    cleaned_records = []
    
    for key, phrases in data.items():
        if not phrases:
            continue
            
        # Join all phrases for this record
        joined_text = " ".join([
            re.sub(r'[^\w\s\.\,\'\"\-\u0e00-\u0e7f]', ' ', phrase)
            for phrase in phrases
        ])
        
        # Basic cleaning
        joined_text = re.sub(r'\s+', ' ', joined_text).strip()
        
        if joined_text:
            cleaned_records.append({
                "id": key,
                "text": joined_text,
                # You'll need to add sentiment labels if you have them
                # "sentiment": ...
            })
    
    return pd.DataFrame(cleaned_records)

# Prepare the data
sentiment_df = prepare_sentiment_data(raw_data)

# Basic data analysis - helpful for sentiment model preparation
print(f"Total records: {len(sentiment_df)}")

# Example feature extraction for text classification
# This creates a bag-of-words representation
vectorizer = CountVectorizer(max_features=100)
X = vectorizer.fit_transform(sentiment_df['text'])

# Get the most common words
word_counts = dict(zip(vectorizer.get_feature_names_out(), X.sum(axis=0).tolist()[0]))
top_words = dict(sorted(word_counts.items(), key=lambda x: x[1], reverse=True)[:20])

print("\nTop 20 words in the dataset:")
for word, count in top_words.items():
    print(f"{word}: {count}")

# Analyze text length - important for model selection
sentiment_df['text_length'] = sentiment_df['text'].apply(len)
print(f"\nAverage text length: {sentiment_df['text_length'].mean():.2f} characters")
print(f"Min text length: {sentiment_df['text_length'].min()} characters")
print(f"Max text length: {sentiment_df['text_length'].max()} characters")

# Detect languages for separate model training if needed
def detect_language(text):
    # Simple detection based on Thai character presence
    if re.search(r'[\u0e00-\u0e7f]', text):
        return 'thai'
    return 'other'  # Mostly English or numbers

sentiment_df['language'] = sentiment_df['text'].apply(detect_language)
language_counts = sentiment_df['language'].value_counts()
print("\nLanguage distribution:")
print(language_counts)

# Extract common patterns that might be useful for sentiment analysis
# For example, price mentions might correlate with sentiment
def extract_price_mentions(text):
    # Look for number patterns followed by common price indicators
    price_patterns = re.findall(r'\d+[\,\.]?\d*\s*(?:บาท|baht|\.|\-)', text.lower())
    return len(price_patterns) > 0

sentiment_df['has_price'] = sentiment_df['text'].apply(extract_price_mentions)
price_counts = sentiment_df['has_price'].value_counts()
print("\nRecords mentioning prices:")
print(price_counts)

# Extract potential sentiment indicators
def extract_sentiment_words(text):
    # This is a very simple example - you'd use proper sentiment lexicons
    positive_words = ['ดี', 'special', 'free', 'new', 'best', 'ultimate']
    negative_words = ['ไม่', 'no', 'not']
    
    text_lower = text.lower()
    pos_count = sum(1 for word in positive_words if word.lower() in text_lower)
    neg_count = sum(1 for word in negative_words if word.lower() in text_lower)
    
    return pos_count, neg_count

sentiment_df[['positive_words', 'negative_words']] = pd.DataFrame(
    sentiment_df['text'].apply(extract_sentiment_words).tolist(), 
    index=sentiment_df.index
)

print("\nPotential sentiment indicators:")
print(f"Records with positive words: {(sentiment_df['positive_words'] > 0).sum()}")
print(f"Records with negative words: {(sentiment_df['negative_words'] > 0).sum()}")

# Prepare for modeling
print("\nData is ready for sentiment modeling!")
print("Next steps would be:")
print("1. Label a sample of the data with sentiment (if not already labeled)")
print("2. Split into training/validation/test sets")
print("3. Apply appropriate NLP preprocessing (tokenization, etc.)")
print("4. Train sentiment models (consider separate models for Thai and non-Thai text)")
print("5. Evaluate model performance")

# Return the prepared DataFrame
sentiment_df.head(10)
```
